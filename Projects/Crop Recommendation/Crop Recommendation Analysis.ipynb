# Univariate Graphical Analysis

```python
df.columns
```

```python
# To check or appoint as numerical or categorical 
num = []
cat = []
for i in df.columns:
    if df[i].dtype == "object":
        cat.append(i)
    else:
        num.append(i)
```

```python
num
```

```python
cat
```

```python
# to use seaborn for data visualisation....
# we checking for the density/frequency of the columns (the destribution)
# added plt.show() so that it will show all instead of just first one or last one
for i in num:
    sns.kdeplot(data = df, x = i)
    plt.show()
```

```python
for i in num:
    sns.histplot(data = df, x = i)
    plt.show()
```

```python
for i in num:
    sns.boxplot(data = df, x = i)
    plt.show()
```

```python
# For categorical (you don't need for loop since its only label column)
sns.countplot(data = df,y = "label") 
plt.show()
```
