Above shows the univariate statistic of the columns

```python
# To check or appoint as numerical or categorical 
num = []
cat = []
for i in df.columns:
    if df[i].dtypes == "object":
        cat.append(i)
    else:
        num.append(i)
```

```python
num
```

```python
cat
```

# Graphical Univariate Analysis

```python
# Graphical Analysis for univariate analysis for num
for i in num:
    sns.kdeplot(data = df, x = i)
    plt.show()
```

```python
for i in num:
    sns.histplot(data = df, x = i)
    plt.show()
```

```python
for i in num:
    sns.boxplot(data = df, x = i)
    plt.show()
```

```python
#Categorical graphs
for i in cat:
    sns.countplot(data = df, x = i)
    plt.show()
```
